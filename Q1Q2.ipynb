```python
data_dir_path='/content/drive/My Drive/Data/DSA2/'

import lxml.etree

import json


# receives the extract result from the spider and prints out the content
class ConsoleWriterPipeline(object):
    def open_spider(self, spider):
        None
    def close_spdier(self, spider):
        None
    
    def process_item(self, item, spider):
        line = json.dumps(dict(item)) + "\n"
        print(line)
        return item
    
# receives the extract result from the spider and appends them into a JSON Line file, (each line is a json)
class JsonWriterPipeline(object):
    def open_spider(self, spider):
        self.file = open(data_dir_path+'result.json', 'w')

    def close_spider(self, spider):
        print('JSON File Generated')
        self.file.close()

    def process_item(self, item, spider):
        line = json.dumps(dict(item)) + "\n"
        self.file.write(line)
        return item

    
```

```python
import logging
import scrapy
from scrapy.crawler import CrawlerProcess

class QuotesSpider(scrapy.Spider):
    name = "quotes"
    start_urls = [
        'https://webscraper.io/test-sites/e-commerce/allinone/computers/tablets'
    ]
    custom_settings = {
        'LOG_LEVEL': logging.WARNING,                            # Default : Debug
        'ITEM_PIPELINES': {'__main__.JsonWriterPipeline': 1}  # Used for pipeline
    }
    
    def parse(self, response):
        for quote in response.css('div.row'):
            yield {
                'product': quote.css('h4 a.title::text').getall(),
                'description': quote.css('div.caption p.description::text').getall(),
                'price': quote.css('div.caption h4.pull-right::text').getall(),
                'review': quote.css('div.ratings p.pull-right::text').getall()
            }
    
```

```python
quotes_crawler_process = CrawlerProcess({
    'USER_AGENT': 'Mozilla/5.0 (compatible; MSIE 7.0; Windows NT 5.1)'
})

quotes_crawler_process.crawl(QuotesSpider)
quotes_crawler_process.start()
# '''
```

```python
import json

jsonObjects = []
with open('/content/drive/MyDrive/Data/DSA2/result.json') as f:
    for jsonObj in f:
        jsonDict = json.loads(jsonObj)
        jsonObjects.append(jsonDict)

print(jsonObjects[1])
```

```python
products_list = []
info_dict = jsonObjects[1]
for i in range(len(info_dict['product'])):
  temp_dict = {}
  temp_dict['product'] = info_dict['product'][i]
  temp_dict['description'] = info_dict['description'][i]
  temp_dict['price'] =info_dict['price'][i]
  temp_dict['review'] = info_dict['review'][i]
  products_list.append(temp_dict)
```

```python
print(len(products_list))
products_list
```

```python
def SearchbyReviewGreaterthan(int_review):
  return_result = []

  for i in products_list:
    if int(i['review'].split(sep=' ')[0]) > int_review:
      return_result.append(i)

  return(return_result)


SearchbyReviewGreaterthan(13)
```
