```python
import pandas as pd
```

```python
data_dir_path='/content/drive/MyDrive/'
```

```python
import xml.etree.ElementTree as ET
import csv
 
root_node = ET.parse(data_dir_path+"catalogQ3.xml").getroot()

artist_title = []

for tag in root_node.findall('Album'):
    artist = tag.attrib['artist']
    print(artist)
    title = tag.attrib['title']
    print(title)
    artist_title.append(artist)
    artist_title.append(title)

count = 1

with open('tracks.csv', 'w', newline='') as file:
  writer = csv.writer(file)
  writer.writerow(['artist','title','rating','track'])

  for tag in root_node.findall('Album/Track'):
      rating = tag.attrib['rating']
      texts = tag.text
      if count < 7:
        temporary1 = [artist_title[0], artist_title[1], rating, texts]
      else:
        temporary1 = [artist_title[2], artist_title[3], rating, texts]
      
      writer.writerow(temporary1)
      print(temporary1)
      count += 1
```

```python
from statsmodels.stats import weightstats as stests
ztest ,pval1 = stests.ztest(df22.loc[df22.earnings > 100000, :]['hours'], df22.loc[df22.earnings < 100000, :]['hours'], alternative='larger')
print(float(pval1))

if(pval1<0.05):
    print('We reject null hypothesis')
else:
    print('We fail to reject null hypothesis')
```
